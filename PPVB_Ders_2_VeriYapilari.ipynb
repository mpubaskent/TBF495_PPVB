# Veri Yapıları

#### Tuple
Tuple türü veri yapıları, değerleri değiştirilemeyen, sıralanmış elemanlar (nesneler) içerirler.
İndeks numaraları olup bunlarla ilgili elemana erişmek mümkündür.

```python
musteri_adlari   = ("Ahmet", "Bekir", "Cemal", "Davut")
print (musteri_adlari)

sipariş_nolari   = (1969, 1983, 1991, 2014, 2018)
print (sipariş_nolari)

sipariş_bilgisi  = ("Murat Paşa Uysal","Adres: B.Ünv.",2.5,"15/O2/2021","5XX XXX 69 91")
print (sipariş_bilgisi)

sipariş_bilgisi  = (musteri_adlari, sipariş_nolari)
print (sipariş_bilgisi)
```

```python
# Tuple'ın bir eleman değeri değiştirilemez!
# musteri_adlari [2] = "Ekrem"
```

```python
# Tuple'ın sonuna bir değerin eklenmesi

musteri_adlari  = ("Ahmet", "Bekir", "Cemal", "Davut")
print (musteri_adlari)
musteri_adlari  = musteri_adlari + ("Faruk",)
print (musteri_adlari)
musteri_adlari += ("Galip",)
print (musteri_adlari)
```

```python
# Tuple elemanları arasında gezinme
musteri_adlari  = ("Ahmet", "Bekir", "Cemal", "Davut")

for deger in musteri_adlari:
    print ("Değer :", deger)

print("--------------------------")  

indeks = 0
for item in musteri_adlari:
    print ("İndeks Nu:", indeks, "  Değer:", musteri_adlari[indeks])
    indeks += 1
```

```python
#Tuple'da bir değerin tekrar sayısının bulunması
musteri_adlari   = ("Ahmet", "Bekir", "Cemal", "Davut", "Ekrem")
musteri_adlari.count("Ahmet")
musteri_adlari = musteri_adlari  + ("Ahmet", "Davut")
print (musteri_adlari)
print ("Ahmet ismi ", musteri_adlari.count("Ahmet"), "defa tekrar etmektedir")
```

```python
# İç içe Tuple'lar
tuple1 = (10, 20, 30, 40, 50)
print ("Tuple1 :", tuple1)
print ("Tuple1 in 2nci elemanı: ",tuple1[2])
print("-----------------------------------")
tuple2 = (tuple1, 60, 70, 80, 0,100)
print ("Tuple2 :", tuple2)
print ("Tuple2 in 2nci elemanı: ",tuple2[2])
print ("Tuple2 in 0ncı elemanı: ",tuple2[0])
print("-----------------------------------")
tuple3 = (110,120,130,140,150 ,tuple2)
print ("Tuple3 :", tuple3)
print ("Tuple3 in 2nci elemanı: ",tuple3[2])
print ("Tuple3 in 5nci elemanı: ",tuple3[5])
print ("Tuple3 in 5nci elemanın 0ncı elemanı: ",tuple3[5][0])
print ("Tuple3 in 5nci elemanın 0ncı elemanın 3ncü elemanı: ",tuple3[5][0][3])
```

```python
# List türü veri yapısının yaratılması
musteri_adlari = []
print(musteri_adlari)
print("--------------------------") 
musteri_adlari  = ["Ayşe", "Beyza", "Cemile", "Emine", "Elif"]
print(musteri_adlari)
print("--------------------------") 
print(musteri_adlari[:2]) # ikinci elemana kadar
print("--------------------------") 
print(musteri_adlari[2:]) # ikinci eleman dahil ondan sonrası
```

```python
musteriler = ["Ayşe", "Beyza", "Cemile", "Emine","Fatma"]
print(musteriler)
#Yeni elemanın araya eklenmesi
musteriler.insert(2, "Billur")
print(musteriler)
# Belirli indeksteki elemanın güncellenmesi
musteriler[3] = "Didem"
print(musteriler)
# Belirli indeksteki elemanın silinmesi
del musteriler[1]
print(musteriler)
# Listenin sonundakii elemanın silinmesi
musteriler.pop()
print(musteriler)
# Belirli indeksteki elemanın silinmesi
musteriler.pop(1)
print(musteriler)
# Belirli bir değer sahip elemanın silinmesi
musteriler.remove("Didem")
print(musteriler)
```

#### List İçindeki Verinin Tekrar Sayısı

```python
from collections import Counter

liste = [1, 2, 3, 4, 1, 2, 3, 1, 2, 1, 5]
tekrar_sayilari = Counter(liste)
print(tekrar_sayilari)
print("----------------------------------")

for cift in tekrar_sayilari.items():
    print("Değer:", cift[0], "'in' Tekrar Sayısı: ", cift[1])
   
```
