###  "if ", " İf…else ", " if…elif "  Deyimlerinin Kullanımları



```python
# if..else Deyimin Kullanımı
sart = True 
if (sart): 
    print ("True ise işlemler yap") 
else:
    print ("False ise işlemler yap")
    
```

```python
# if..else Deyimin Kullanımı
sart = False 
if (sart):
    print ("True ise işlemler yap") 
else:
    print ("False ise işlemler yap")
```

```python
# Kullanıcı alışveriş ve siparişini tamamlamıştır
siparisDurumu = True 
if (siparisDurumu):
    print ("Üye müşteri işlemleri ekranına yönlendir.") 
    print ("Üye müşteri bilgilerini veri tabanından getir.") 
    print ("Üye müşteri onayını alarak ödeme işlemleri başlat.")
else:
    print ("Kullanıcı bilgileri ekranına yönlendir.")
    print ("Teslimat adresi ve irtibat bilgilerini iste.")
    print ("Üye olmak isterse üyelik işlemlerini gerçekleştir.")
    print ("Kullanıcı veya Üye müşteri onayını alarak ödeme işlemleri başlat.")
```

#### Durum:
Bir üye müşteri çevrimiçi (online) alışveriş sistemine giriş yapmış, alışveriş sepetini ürünlerle doldurmuş ve siparişini oluşturmuştur.  Ödeme işlemlerini gerçekleştirmek istemektedir.

```python
# Üye müşteri siparişi ve ödeme işlemleri
siparisBittiMi = True 
kullanciUyeMi  = False
if (siparisBittiMi):
    if (kullanciUyeMi):
       print ("Üye müşteri ödeme işlemlerini gerçekleştir.") 
    else:
       print ("Kullanıcı ödeme işlemlerini gerçekleştir.")
```

#### Durum:
Çevrimiçi (online) alışveriş sistemi indirim haftası düzenlemektedir. Sipariş tutarlarına göre farklı oranlarda indirim yapmaktadır.Bu kapsamda 500 TL’e kadar %5; 500-1500 TL arası %8 ve 1500-2000 TL arasında %10 indirim uygulamaktadır.

```python
# Üye kullanıcılara farklı oranlarda indirim uygulanması
siparisTutari = 2500
odemeTutari   = 2500
indirimOrani  = 0.0
if (siparisTutari < 500):
    indirimOrani = 0.05
    odemeTutari  = siparisTutari - (siparisTutari * indirimOrani)
    print ("%5 indirim yapılmış olup ödeme tutarı = ", odemeTutari)
elif (siparisTutari >= 500) and (siparisTutari < 1500):
    indirimOrani = 0.08
    odemeTutari  = siparisTutari - (siparisTutari * indirimOrani)
    print ("%8 indirim yapılmış olup ödeme tutarı = ", odemeTutari)
elif (siparisTutari >= 1500) and (siparisTutari <= 2500):
    indirimOrani = 0.10
    odemeTutari  = siparisTutari - (siparisTutari * indirimOrani)
    print ("%10 indirim yapılmış olup ödeme tutarı = ", odemeTutari)
```

### "for loop"  ve " while loop" Deyimlerinin Kullanımları
#### Sabit Sayıda "for loop" Deyimi

```python
siparisSayisi = int (input("Sipariş Sayısını Giriniz :"))
for sayac in range(siparisSayisi):
    print (sayac)
```

#### İki Sayı Aralığı "for loop" Deyimi

```python
baslangicSiparisNu = int (input("Başlangıç Sipariş Numarasını Giriniz :"))
bitisSiparisNu     = int (input("Bitiş Sipariş Numarasını Giriniz :"))
for sayac in range(baslangicSiparisNu, bitisSiparisNu):
    print (sayac)
```

#### İki Sayı Aralığı ve Adım Sayısı "for loop" Deyimi

```python
baslangicSiparisNu = int (input("Başlangıç Sipariş Numarasını Giriniz :"))
bitisSiparisNu     = int (input("Bitiş Sipariş Numarasını Giriniz :"))
adimSayisi         = int (input("Adım Sayısını Giriniz :"))
for sayac in range(baslangicSiparisNu, bitisSiparisNu, adimSayisi):
    print (sayac)
```

#### İki Sayı Aralığı ve Geriye Sayma "for loop" Deyimi

#### String Değer Dizisi "for loop" Deyimi

```python
isimListesi = ["Ahmet","Bekir","Cemal","Davut","Ekrem", "Faruk"]
arananIsim  = input("Aranan İsmi Giriniz : ")
for isim in isimListesi:
    if (isim == arananIsim):
        print("Aradığınız İsim Bulunmuştur --> ", isim)
```

#### String Değer Dizisi "for loop else" Deyimi

```python
isimListesi = ["Ahmet","Bekir","Cemal","Davut","Ekrem", "Faruk"]
arananIsim  = input("Aranan İsmi Giriniz : ")
for isim in isimListesi:
    if (isim == arananIsim):
        print("Aradığınız İsim Bulunmuştur --> ", isim)
else:
    print("Aradığınız İsim Bulunamamıştır !!!")
```

#### String Değer Dizisi "for loop pass" Deyimi

```python
isimListesi = ["Ahmet","Bekir","Cemal","Davut","Ekrem", "Faruk"]
arananIsim  = input("Aranan İsmi Giriniz : ")
for isim in isimListesi:
    if (isim == arananIsim):
        print("Aradığınız İsim Bulunmuştur --> ", isim)
        pass
print ("Döngü bitmiştir")
```

#### İç İçe "for..loop" Deyimleri

```python
for satir in range (7):
    for sutun in range (satir):
        print (sutun, end='')
    print()
    
```

#### "while..loop" Deyimleri

```python
sayac= 1
tekrarSayisi = int(input("Döngü Tekrar Sayısını Giriniz :"))
while sayac <= tekrarSayisi:
    print(sayac) 
    sayac +=1
```

#### İç İçe "while..loop" Deyimleri

```python
sayac1= 1
sayac2= 1
tekrarSayisi = int(input("Döngü Tekrar Sayısını Giriniz :"))
while sayac1 <= tekrarSayisi:
    while sayac2 <= sayac1:
        print(sayac2, end='') 
        sayac2 +=1
    print()
    sayac2 = 1
    sayac1 +=1
```

### İstisna Yönetimi (Exception Handling)

Programlamada istisnai durumların (beklenmeyen durumların) (exception) farklı kaynakları olabilmektedir. Bunların birkaçı:
    Girdi çıktı işlemleri,
    Matematiksel hesaplamalar,
    Dosya işlemleri,
    Mantıksal kodlama hataları vb.’leridir.
Bu durumlarda; meydana gelen olayın :
    Ele alınması (exception handling), 
    Kayıt altına alınması (logging) ve
    Programın sorunsuz biçimde çalışması sağlanmalıdır.
İstisna Yönetimine yönelik kodlar bu gereksinimleri karşılamak amacıyla gerçekleştirilmektedir.

```python
musteriNo = int(input("Müşteri Numarasını Giriniz :"))
print (musteriNo," Nolu Müşteri Bilgisini Girdiniz")
```

```python
try:
    musteriNo = int(input("Müşteri Numarasını Giriniz :"))
except ValueError:
    print ("HATA: Müşteri Numarasını Yanlış Formatta Girdiniz!!!")
else:
    print ("    Try bloğundan sonraki deyimler yerine getirilirler...")
    print ("   ",musteriNo," Nolu Müşteri Bilgisini Girdiniz")
```
